# 1. Plots for Categorical Variables

## I. Categorical Plots (catplot) for a Single Categorical Variable

*All categorical plots below have been created using built-in 'tips' dataset from the Seaborn library.*

### *I. Categorical Distribution Plots*

Categorical Distribution Plots are used to visualize the distribution of a numerical variable across the categories of a single categorical variable.

**Box and Whisker Plot**

A box plot describes the distribution of a continuous variable by plotting the summary of five statistical terms: minimum, maximum, second quartile (50th percentile), first quartile (25th percentile), and third quartile (75th percentile) in the dataset.

```python
# Method I
sns.catplot(data=tip, x='day', y='total_bill', kind='box')
```

```python
# Method II
sns.boxplot(data=tip, x='day', y='total_bill')
```

**Violin Plot**

A violin plot combines the features of a box plot and a kernel density plot, providing a deeper insight into the distribution of a numerical variable across different categories of a categorical variable.

```python
# Method I
sns.catplot(data=tip, x='day', y='total_bill', kind='violin')
```

```python
# Method II
sns.violinplot(data=tip, x='day', y='total_bill')
```

**Boxen Plot (or, Letter Value Plot)**

*- A boxen plot reveals the distribution of a numerical variable across different groups of a categorical variable, especially when dealing with large, skewed datasets that have heavy tails.*

*- The boxen plot is similar to a box plot but provides more information about the shape of the distribution, particularly in the tails.*

```python
# Method I
sns.catplot(data=tip, x='total_bill', y='day', kind='boxen')
# A horizontal boxen plot displaying a categorical variable on the y-axis
```

```python
# Method II
sns.boxenplot(data=tip, x='total_bill', y='day')
# A horizontal boxen plot displaying a categorical variable on the y-axis
```

### *II. Categorical Estimate Plots*

Categorical Estimate Plots are used to visualize statistical estimates — such as means or counts — of a numerical variable across the categories of a categorical variable.

**Point Plot**

*- A point plot displays point estimates (e.g., the mean value by default or median) of a numerical variable across the categories of a categorical variable, along with error bars that represent variability, such as confidence intervals or standard deviations.*

*- It displays a point or dot for each category of the categorical variable and connects them with line segments, with error bars representing confidence intervals by default.*

```python
# Method I
sns.catplot(data=tip, x='day', y='total_bill', kind='point') 
# The 95% CI is represented by vertical lines intersecting with points
```

```python
# Method II
sns.pointplot(data=tip, x='day', y='total_bill')
```

**Bar Plot**

A bar plot displays bars that represent a summary statistic (usually the mean) of a numerical variable for each category of a categorical variable. Error bars indicate variability, such as confidence intervals by default or standard deviation.

```python
# Method I
sns.catplot(data=tip, x='day', y='total_bill', kind='bar')
# The 95% confidence interval (CI) is represented by the error bars, i.e. the vertical lines at the top of the bars
```

```python
# Method II
sns.barplot(data=tip, x='day', y='total_bill')
```

**Count Plot**

*- A count plot is a type of bar chart that shows the number of occurrences (or frequency) of each category in a categorical variable.*

*- Unlike a bar plot, which displays a summary statistic such as the mean of a numerical variable, a count plot simply shows how many times each category occurs.*

```python
# Method I
sns.catplot(data=tip, x='day', kind='count')
```

```python
# Method II
sns.countplot(data=tip, x='day')
```

### *III. Categorical Scatter Plots*

Categorical Scatter Plots display individual data points of a numerical variable across the categories of a categorical variable. These plots are especially useful when the goal is to visualize raw data points rather than summary statistics such as means or medians.

**Strip Plot**

A strip plot is a type of categorical scatter plot that displays individual data points for a numerical variable across the different levels of a categorical variable.

```python
# Method I
sns.catplot(data=tip, x='day', y='total_bill', kind='strip') # By default, catplot is a 'strip' plot
```

```python
# Method II
sns.stripplot(data=tip, x='day', y='total_bill')
```

**Swarm Plot**

*- A swarm plot is a type of categorical scatter plot that displays individual data points for a numerical variable while avoiding overlap, making it easier to visualize how values are distributed across the levels of a categorical variable.*

*- It’s often called a “beeswarm” plot because the points are arranged like a swarm of bees — neatly packed but not stacked on top of each other.*

```python
# Method I
sns.catplot(data=tip, x='day', y='total_bill', kind='swarm')
```

```python
# Method II
sns.swarmplot(data=tip, x='day', y='total_bill')
```

## II. Categorical Plots for Multiple Categorical Variables


### *Facet Grid*

*- In Seaborn, FacetGrid itself is a layout tool. It is not inherently a multivariate plot, but it enables multivariate visualization.*

*- It's a plotting framework that creates a grid of subplots based on one or more categorical variables.*

*- We can map a plotting function (like sns.scatterplot) onto each subplot.*

*- If we map a function that shows relationships between two or more variables (e.g., x, y, and hue), and then facet by additional variables (row, col), we are effectively visualizing multiple dimensions of our data.*

The Facet Grid below has been created using built-in 'tips' dataset from the Seaborn library.

```python
tip.head()
```

```python
# Creating a grid of subplots based on the categorical variables (i.e., time, sex, and smoker)
f_g = sns.FacetGrid(data=tip, row='time', col='sex', hue='smoker')
# Mapping a plotting function (sns.scatterplot) onto each subplot that shows relationships between two or more variables (e.g., x, y, and hue)
f_g.map(sns.scatterplot, 'total_bill', 'tip').add_legend()
```

# 2. Plots for Continuous Variables

## I. Univariate Distribution Plots 

*All univariate distribution plots below have been created using built-in 'tips' dataset from the Seaborn library.*

**Histogram**

*- Histogram is primarily used to display the distribution of a single continuous variable.*

*- A histogram is a type of bar chart that displays the distribution of a numerical dataset by grouping values into intervals (called 'bins') and showing how many data points fall within each bin.*


```python
# Method I
sns.displot(data=tip, x='total_bill', kind='hist')
```

```python
# Method II
sns.histplot(data=tip, x='total_bill', bins=20)
```

**Kernel Density Estimate (KDE) Plot**

*Displays the probability density function of a continuous variable.*

```python
# Method I
sns.displot(data=tip, x='total_bill', kind='kde')
```

```python
# Method II
sns.kdeplot(data=tip, x='total_bill')
```

**Empirical Cumulative Distribution Function (eCDF) plot**

*- Used for a single numerical or quantitative variable.*

*- Displays the eCDF as the proportion (or percentage) of observations that are less than or equal to each data value.*

```python
# Method I
sns.displot(data=tip, x='total_bill', kind='ecdf')
```

```python
# Method II
sns.ecdfplot(data=tip, x='total_bill')
```

**Rug Plot**

*- Displays the distribution of a single numerical variable, most commonly the continuous variable.*

*- Visualizes the distribution of data as marks or ticks along an x-axis.*

```python
# RUG plot with KDE
sns.displot(data=tip, x='total_bill', kind='kde', rug=True)
```

```python
# Rug Plot
sns.rugplot(data=tip, x='total_bill')
```

## II. Bivariate Plots 

### *I. Relational Plots*

Relational Plots are used to show the relationship between two numerical variables.

*The two relational plots below (scatter plot and line plot) have been created using built-in 'tips' dataset from the Seaborn library.*

**Scatter Plot**

*- Shows the relationship between two continuous variables.*

*- Display symbols at the X, Y coordinates of the data points for the paired variables.*

```python
sns.scatterplot(data=tip, x='total_bill', y='tip', s=100) # Here, 's' is the size arguement
```

**Line Plot**

*- Visualizes the continuous data where the data points are connected by straight lines.*

*- Displays trends, patterns or relationship in data over time (usually treated as a continuous variable) or across a continuous scale.*

*- Also useful to show a trend or progression even if the data is technically discrete (like group sizes: 5, 10, 15)*

### *II. Regression Plots*

*- Regression plots in Seaborn are used to visualize the relationship between two continuous variables — usually an independent (predictor) variable and a dependent (response) variable — by fitting a regression line through the data.*

*- Regression plots show (1) a scatter plot of the data points, (2) a regression line that models the relationship (usually linear), and (3) an optional confidence interval around the regression line to show variability.*

*The regplot and lmplot below have been created using built-in 'iris' dataset from the Seaborn library.*

**Regression Plot (regplot)**

Regplot is used for visualizing linear relationships between two variables. It’s like a scatterplot with a built-in regression line, perfect for spotting trends and correlations.

```python
# Loading the built-in 'iris' dataset from the Seaborn library
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.regplot(data=iris, x='petal_length', y='petal_width')
```

**Linear Model Plot (lmplot)**

Lmplot is a high-level plotting function designed to visualize linear relationships between two variables, much like regplot(), but with added flexibility for faceting and categorical grouping. This is a more powerful, multi-dimensional version of regplot()

```python
# Using 'iris' dataset
sns.lmplot(data=iris, x='petal_width', y='sepal_width')
```

```python
# Using 'tips' dataset with the categorical variable (i.e. 'sex' column) grouped by 'hue' (color)
sns.lmplot(data=tip, x='total_bill', y='tip', hue='sex')
```

### *III. Bivariate Heatmap*

**Bivariate Heatmap (Matrix Plot)**

*- A bivariate heatmap is a two-dimensional visualization technique that illustrates the relationship between two variables using color intensity or gradients.*

*- One axis represents Variable A, the other represents Variable B, and the color of each cell indicates the density, frequency, or magnitude of observations corresponding to that variable pair.*

*- While the axes (month, year) are categorical (although 'year' is numerical in nature, it's often used to group or label data and is therefore treated as a categorical variable), **the heatmap’s primary purpose is to visualize the continuous variable** (passengers) using color intensity.*

*- Heatmap is one of the primary ways to visualize matrix-style data. For it to work effectively, our dataset should already be structured in matrix form.*

*The heatmap below has been created using built-in 'flights' dataset from the Seaborn library.*

```python
# Loading built-in 'flights' dataset from the Seaborn library
flight = sns.load_dataset('flights')
flight.head()
```

```python
# Reshaping data into matrix form using the pivot_table() function
fl_pivot = flight.pivot_table(values='passengers', index='month', columns='year')
fl_pivot
```

```python
sns.heatmap(fl_pivot, cmap='magma', linecolor='white', linewidths=2)
```

## III. Multivariate Plots 

### *I. Cluster Map (Multivariate Matrix Plot)*

*- A clustermap is a powerful visualization tool that combines a heatmap with hierarchical clustering to reveal patterns and relationships in multivariate datasets.*

*- It uses hierarchical clustering to group similar rows and columns.*

*- A clustermap is a multivariate plot because it analyzes similarity patterns across multiple variables simultaneously. For example, in the plot below, each row (representing a month) is compared across all years, while each column (representing a year) is compared across all months. The standard_scale=1 argument standardizes the data across columns, further highlighting its multivariate nature.*

*The clustermap below has been created using built-in 'flights' dataset from the Seaborn library.*

```python
# Loading built-in 'flights' dataset from the Seaborn library
flight = sns.load_dataset('flights')
flight.head()
```

```python
# Reshaping data into matrix form using the pivot_table() function
fl_pivot = flight.pivot_table(values='passengers', index='month', columns='year')
fl_pivot
```

```python
sns.clustermap(fl_pivot, cmap='coolwarm', standard_scale=1)
```

 ### *II. Pair Plot (pairplot)*

*- A pair plot creates a grid of scatter plots, where each subplot displays the relationship between two numerical variables, using a consistent scale for each variable across all plots.*

*- The diagonal typically shows univariate distributions (like histograms or KDE plots), while the off-diagonal plots show bivariate relationships.*

*The Pairplot below has been created using built-in 'iris' dataset from the Seaborn library.*

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

### *III. Pair Grid Plot*

*- A PairGrid plot is a highly customizable tool for visualizing pairwise relationships among multiple variables in a dataset.*

*- It creates a grid of subplots, where each subplot shows the relationship between two variables.*

*- It's possible to choose  different plot types for the diagnol, upper triangle and lower triangle.*

*The Pair Grid below has been created using built-in 'iris' dataset from the Seaborn library.*

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Creating a grid of subplots, where each subplot shows the relationship between a pair of variables, and mapping a scatter plot onto each subplot
p_g = sns.PairGrid(iris)
p_g = p_g.map(plt.scatter)
```

```python
# Mapping scatter plot to the upper triangle, histogram to the entire diagonal, and kde plot to the lower triangle
p_g = sns.PairGrid(iris)
p_g.map_upper(plt.scatter)
p_g.map_diag(plt.hist)
p_g.map_lower(sns.kdeplot)
```
